```python
# merge of both dataframes based on ID
df = pd.merge(df1, df2, on='ID')
```

```python
# Convert data types of below attributes to object
convert_col = ['CreditCard','InternetBanking','FixedDepositAccount','Security','Level','HiddenScore']
df[convert_col] = df[convert_col].astype(object)
```

```python
fig, axes = plt.subplots(14,3,figsize=(20,30)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=1, wspace=0.3)

i=0
for feature in df.columns:
    sns.boxplot(data=df,x=feature,ax=axes[i,0],color='red')
    sns.histplot(data=df,x=feature,kde=True,ax=axes[i,1],color='green')
    sns.countplot(x=feature,ax=axes[i,2], data=df)
    
    axes[i, 0].set_title('{} Box Plot'.format(feature))
    axes[i, 1].set_title('{} Hist Plot'.format(feature))
    axes[i, 2].set_title('{} Count Plot'.format(feature))
    
    i += 1
```

```python
fig, axes = plt.subplots(3,3,figsize=(18,18)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3) # required height_space and width_space

#  Function that will accept a DataFrame and axes as input and return pie-charts for the appropriate categorical features
#  Function returns pie charts with % Distribution of the categorical features
def fun(df,axes):
    categorical_cols = df.select_dtypes(include=['object', 'int64', 'float64']).columns
    categorical_cols = categorical_cols.difference(['ID', 'Age', 'CustomerSince', 'HighestSpend', 'ZipCode','MonthlyAverageSpend', 'Mortgage'])

    i=j=0
    for column in categorical_cols:
        if j > 2:
            j = 0
            i += 1
            
        values = df[column].value_counts()
        axes[i, j].pie(values, labels=values.index, autopct='%1.1f%%')
        axes[i, j].set_title('% Distribution of {}'.format(column))

        j+=1

fun(df,axes)
```

```python
def myscatterplot(df, x_column, y_column,hue="",palette=""):
    if hue is not None:
        sns.scatterplot(data=df, x=x_column, y=y_column, hue=hue, palette=palette)
    else:
        sns.scatterplot(data=df, x=x_column, y=y_column)
    
    plt.title(f"Scatterplot of {x_column} vs {y_column}")
    plt.xlabel(f"{x_column}")
    plt.ylabel(f"{y_column}")
    plt.show()

myscatterplot(df,'MonthlyAverageSpend','HighestSpend', hue='LoanOnCard', palette=['red', 'blue'])
```
